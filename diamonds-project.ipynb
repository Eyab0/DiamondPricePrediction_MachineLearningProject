### 4.3 Drop the OutLiers

```python
#diamond = diamond[(diamond["depth"]<75)&(diamond["depth"]>45)]
#diamond = diamond[(diamond["table"]<80)&(diamond["table"]>40)]
#diamond = diamond[(diamond["x"]<30)]
#diamond = diamond[(diamond["y"]<30)]
#diamond = diamond[(diamond["z"]<30)&(diamond["z"]>2)]
#diamond.shape #12 element deleted
```

```python
Q1 = diamond['depth'].quantile(0.25)
Q3 = diamond['depth'].quantile(0.75)
IQR = Q3- Q1
idx = ~((diamond['depth']<(Q1-1.5*IQR))|(diamond['depth']>(Q3+1.5*IQR)))
diamond = diamond[idx] 
diamond.head()
```

```python
Q2 = diamond['table'].quantile(0.25)
Q4 = diamond['table'].quantile(0.75)
IQR1 = Q4- Q2
idx1 = ~((diamond['table']<(Q2-1.5*IQR1))|(diamond['table']>(Q4+1.5*IQR1)))
diamond = diamond[idx1] 
diamond.head()
```

```python
Q5 = diamond['x'].quantile(0.25)
Q6 = diamond['x'].quantile(0.75)
IQR2 = Q6- Q5
idx2 = ~((diamond['x']<(Q5-1.5*IQR2))|(diamond['x']>(Q6+1.5*IQR2)))
diamond = diamond[idx2] 
diamond.head()
```

```python
Q7 = diamond['y'].quantile(0.25)
Q8 = diamond['y'].quantile(0.75)
IQR3 = Q8- Q7
idx3 = ~((diamond['y']<(Q7-1.5*IQR3))|(diamond['y']>(Q8+1.5*IQR3)))
diamond = diamond[idx3] 
diamond.head()
```

```python
Q0 = diamond['z'].quantile(0.25)
Q9 = diamond['z'].quantile(0.75)
IQR4 = Q9- Q0
idx4 = ~((diamond['z']<(Q0-1.5*IQR4))|(diamond['z']>(Q9+1.5*IQR4)))
diamond = diamond[idx4] 
diamond.head()
```

### 6.1 Handling Text and Categorical Attributes



```python
color_cat = diamond[["color"]]
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
color_cat_encoded = ordinal_encoder.fit(color_cat)
color_cat_enc = color_cat_encoded.transform(color_cat)
np.unique(color_cat_enc)
diamond["newcolor"] = color_cat_enc
```

```python
diamond_cat = diamond[["cut","clarity"]]
ordinal_encoder = OrdinalEncoder(categories = [['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], 
                                              ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']])
diamond_cat_encoded = ordinal_encoder.fit(diamond_cat)
diamond_cat_enc= diamond_cat_encoded.transform(diamond_cat)
clarity_cat_encoder = diamond_cat_enc[0:,1]
cut_cat_encoded = diamond_cat_enc[0:,0]
```

```python
diamond["newcut"] = cut_cat_encoded
diamond["newclarity"] = clarity_cat_encoder
```
